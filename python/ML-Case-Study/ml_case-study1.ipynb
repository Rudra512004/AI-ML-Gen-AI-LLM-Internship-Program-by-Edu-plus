```python
# Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order

df["Top-Sellers-Rank"] = df["Quantity"].rank(
    ascending=False,
    method="dense"
)
print(df)

df_sorted = df.sort_values('Quantity', ascending=False)
print(df_sorted[['Product', 'Quantity']])

```

```python
Total_sales=df.groupby('Category')['TotalAmount'].sum()
print(Total_sales)
```

```python
category_summary = (
    df.groupby("Category")
      .agg(
          TotalSales=("TotalAmount", "sum"),
          AverageRating=("Rating", "mean"),
          NumberOfOrders=("OrderID", "count")
      )
      .sort_values("TotalSales", ascending=False)
)

print(category_summary)
```

```python
grouped_summary = (
    df.groupby(["Category", "Product"])
      .agg(
          AvgPrice=("Price", "mean"),
          TotalSales=("TotalAmount", "sum"),
      )
)

print(grouped_summary)
```
