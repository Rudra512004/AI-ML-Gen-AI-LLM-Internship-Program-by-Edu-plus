# Arithmetic Operators and Array in Python
## Using Python `array` Module

### Import Required Module

```python
from array import array
```

### Declare Array (Defined Once)

```python
sales = array('i', [25, 30, 28, 35, 40])
print('Sales Array:', sales)
```

## Problem 1: Total Sales

```python
total_sales = sum(sales)
print('Total Sales:', total_sales)
```

## Problem 2: Average Sales

```python
average_sales = sum(sales) / len(sales)
print('Average Sales:', average_sales)
```

## Problem 3: Highest Sales

```python
highest_sales = max(sales)
print('Highest Sales:', highest_sales)
```

## Problem 4: Lowest Sales

```python
lowest_sales = min(sales)
print('Lowest Sales:', lowest_sales)
```

## Problem 5: Increase Sales by 10%

```python
increased_sales = array('f', [s + s*0.10 for s in sales])
print('Sales after 10% increase:', increased_sales)
```

## Problem 6: Monthly Profit (Expense = 20)

```python
profit = array('i', [s - 20 for s in sales])
print('Monthly Profit:', profit)
```

## Problem 7: Even Sales Values

```python
even_sales = array('i', [s for s in sales if s % 2 == 0])
print('Even Sales:', even_sales)
```

## Problem 8: Square of Each Sales Value

```python
sales_square = array('i', [s**2 for s in sales])
print('Square of Sales:', sales_square)
```

## Problem 9: Sales Above Average

```python
above_average = array('i', [s for s in sales if s > average_sales])
print('Sales Above Average:', above_average)
```

## Problem 10: Discount Using Floor Division

```python
discount = array('i', [s // 5 for s in sales])
print('Discount Values:', discount)
```
