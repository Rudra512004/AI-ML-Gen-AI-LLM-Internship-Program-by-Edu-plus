```python
# question 12. Create a function that accepts text input and returns a list of nouns and verbs separately using POS tagging.
import nltk
from nltk import word_tokenize, pos_tag

def extract_nouns_verbs(text):
    tokens = word_tokenize(text)        
    tagged = pos_tag(tokens)           

    nouns = []
    verbs = []

    for word, tag in tagged:
        if tag.startswith('NN'):       
            nouns.append(word)
        elif tag.startswith('VB'):      
            verbs.append(word)

    return nouns, verbs


text3 = input("Enter Sentence: ")
nouns, verbs = extract_nouns_verbs(text3)

print("Nouns:", nouns)
print("Verbs:", verbs)
```

```python
# question 13 Count the number of adjectives and adverbs present in a given text.
import nltk
from nltk import word_tokenize, pos_tag

def extract_nouns_verbs(text):
    tokens = word_tokenize(text)        
    tagged = pos_tag(tokens)           

    adjectives = []
    adverbs = []

    for word, tag in tagged:
        if tag.startswith('JJ'):       
            adjectives.append(word)
        elif tag.startswith('RB'):      
            adverbs.append(word)

    return adjectives, adverbs


text3 = input("Enter Sentence: ")
adjectives, adverbs = extract_nouns_verbs(text3)

print("Adjectives:", len(adjectives))
print("Adverbs:", len(adverbs))
```

```python
import nltk
from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier
from nltk.tokenize import word_tokenize

documents = [
    (list(movie_reviews.words(fileid)), category)
    for category in movie_reviews.categories()
    for fileid in movie_reviews.fileids(category)
]
def document_features(words):
    return {word: True for word in words}

featuresets = [(document_features(doc), category) for (doc, category) in documents]


classifier = NaiveBayesClassifier.train(featuresets)


test_sentences = [
    "The movie was fantastic and very enjoyable",
    "I hated the film, it was boring",
    "The plot was interesting but slow",
    "What an amazing performance",
    "The movie was terrible"
]


for sentence in test_sentences:
    tokens = word_tokenize(sentence)
    features = document_features(tokens)
    sentiment = classifier.classify(features)
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {sentiment}\n")
```

```python
import nltk
import random
import pickle
from nltk.corpus import movie_reviews, stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk.classify import NaiveBayesClassifier

nltk.download('movie_reviews')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
all_words = FreqDist(w.lower() for w in movie_reviews.words())
top_2000_words = list(all_words)[:2000]

print("Top 10 frequent words:", top_2000_words[:10])
```

```python
def word_exists(review_text, word):
    return word.lower() in word_tokenize(review_text.lower())

print(word_exists("This movie was amazing", "amazing"))
```

```python
def features(words):
    return {word: (word in words) for word in top_2000_words}

documents = [
    (list(movie_reviews.words(fid)), category)
    for category in movie_reviews.categories()
    for fid in movie_reviews.fileids(category)
]

feature_sets = [(features(doc), label) for doc, label in documents]
random.shuffle(feature_sets)

train_set = feature_sets[:1600]
test_set = feature_sets[1600:]
```

```python
classifier = NaiveBayesClassifier.train(train_set)
accuracy = nltk.classify.accuracy(classifier, test_set)
print("Accuracy:", accuracy)
```

```python
classifier.show_most_informative_features(10)
```

```python
def predict_sentiment(sentence):
    words = word_tokenize(sentence.lower())
    feats = features(words)
    return classifier.classify(feats)

print(predict_sentiment("The movie was fantastic"))
print(predict_sentiment("The movie was boring"))
```

```python
# Save
with open("sentiment_model.pkl", "wb") as f:
    pickle.dump(classifier, f)

# Load
with open("sentiment_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

print(loaded_model.classify(features(word_tokenize("Amazing film"))))
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens

print(preprocess_text("This movie was amazingly good and enjoyable"))
```

```python
def sentiment_pipeline(text):
    processed = preprocess_text(text)
    feats = features(processed)
    return classifier.classify(feats)

print(sentiment_pipeline("I really loved this movie"))
print(sentiment_pipeline("This film was terrible"))
```
