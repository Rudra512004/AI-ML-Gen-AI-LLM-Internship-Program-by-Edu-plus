```python
import tensorflow as tf

# Question 1
# Create a scalar tensor with value 10
scalar = tf.constant(10)
print("Q1 Scalar value:", scalar)
print("Q1 Data type:", scalar.dtype)
```

```python
# Question 2
# Create a 1-D tensor (vector)
vector = tf.constant([2, 4, 6, 8])
print("\nQ2 Vector shape:", vector.shape)
print("Q2 Vector numpy value:", vector.numpy())
```

```python
# Question 3
# Create a 2-D tensor (matrix)
matrix = tf.constant([[1, 2], [3, 4]])
print("\nQ3 Matrix shape:", matrix.shape)
print("Q3 Matrix dtype:", matrix.dtype)
```

```python
# Question 4
# Create a tf.Variable
var = tf.Variable(5.0)
print("\nQ4 Variable value:", var.numpy())
```

```python
# Question 5
# Update variable using assign()
var.assign(10.0)
print("\nQ5 Updated variable value:", var.numpy())
```

```python
# Question 6 
var_matrix = tf.Variable([[1.0, 2.0],
                          [3.0, 4.0]])

var_matrix.assign_add(tf.ones_like(var_matrix))

print("Matrix after assign_add:")
print(var_matrix.numpy())
```

```python
# Question 7
# Compute y inside GradientTape
x = tf.Variable(3.0)
with tf.GradientTape() as tape:
    y = x**2 + 2*x + 1

print("\nQ7 x value:", x.numpy())
print("Q7 y value:", y.numpy())
```

```python
# Question 8
# Compute gradient dy/dx
gradient = tape.gradient(y, x)
print("\nQ8 Gradient dy/dx:", gradient.numpy())
```

```python
# Question 9
# Matrix multiplication
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])
matmul_result = tf.matmul(a, b)

print("\nQ9 Matrix multiplication result:")
print(matmul_result.numpy())
```

```python
# Question 10
# Random tensor using tf.random.uniform
random_tensor = tf.random.uniform((3, 3))
print("\nQ10 Random tensor shape:", random_tensor.shape)
print("Q10 Random tensor dtype:", random_tensor.dtype)
print("Q10 Random tensor values:")
print(random_tensor.numpy())
```
