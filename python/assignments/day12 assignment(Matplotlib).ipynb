```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 11)
y = 2*x + 3

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x), linestyle='-', label='sin(x)')
plt.plot(x, np.cos(x), linestyle='--', label='cos(x)')
plt.legend()
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = np.random.randint(50, 200, size=4)

plt.bar(products, sales)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, c=colors)
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.show()
```

```python
fruits = ['Apple', 'Banana', 'Mango', 'Grapes', 'Orange']
sizes = [20, 25, 30, 15, 10]

plt.pie(sizes, labels=fruits, autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)

plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine Wave")

plt.show()
```

```python
x = np.arange(0, 10)
y = x**2

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.savefig("line_plot.png")
plt.show()
```

```python
x = np.arange(1, 6)

plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, x)

plt.subplot(2, 2, 2)
plt.bar(x, x*2)

plt.subplot(2, 2, 3)
plt.scatter(x, x*3)

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(100))

plt.show()
```

```python
x = np.linspace(0, 5, 50)
y = np.exp(x)

plt.plot(x, y, marker='o', linestyle='--')
plt.show()
```
