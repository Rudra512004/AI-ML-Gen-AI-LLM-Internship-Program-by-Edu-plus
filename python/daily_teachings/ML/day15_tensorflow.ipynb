```python
import tensorflow as tf
print(tf.__version__)
```

```python
scalar = tf.constant(42) #0-D tensor(scalar)
vector= tf.constant([1,2,3,4]) #1-D tensor (vector)
matrix = tf.constant([[1,2],[3,4]]) #2-D tensor (matrix)
tensor_3d= tf.constant([[[1,2],[3,4],[5,6],[7,8]]]) #3-D tensor
```

```python
print("\nTensor Example: ")
print(f"Scalar shape: {scalar.shape},value:{scalar.numpy()}")
print(f"Vector shape: {scalar.shape},value:{vector.numpy()}")
print(f"Matrix shape: {matrix.shape}")
print(f"Matrix shape: {tensor_3d.shape}")
```

```python
#Tensor properties
print(f"\nTensor dtype: {matrix.dtype}")
print(f"Tensor rank: {tf.rank(matrix).numpy()}")
print(f"Tensor size: {tf.size(matrix).numpy()}")
```

```python
#common mathematical functions 
x=tf.constant([1.0,2.0,3.0])
print(f"\nSquare: {tf.square(x).numpy()}")
print(f"Squareroot: {tf.sqrt(x).numpy()}")
print(f"Exponential: {tf.exp(x).numpy()}")

#creating varialbles
weights=tf.Variable([[1.0,2.0],[3.0,4.0]],name="weights")
bias= tf.Variable([0.5,0.5],name="bias")
print("\nVariables Example:")
print(f"Initial weights:\n{weights.numpy()}")


#updating variables
weights.assign([[2.0,3.0],[4.0,5.0]])
print(f"Updated weights:\n{weights.numpy()}")
```
