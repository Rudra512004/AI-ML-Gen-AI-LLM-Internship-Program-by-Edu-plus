```python
#Create two class Dog and Cat, both having a method sound(). write a function that calls sound() for both objects demostrating polymorphism.
class Dog:
    def sound(self):
        print("Bark")
class Cat:
    def sound(self):
        print("Meow")
def make_sound(animal):
    animal.sound()
dog = Dog()
cat = Cat()
make_sound(dog)
make_sound(cat)
```
