# Arithmetic Operators and Array in Python
## Complete Array Operations with Examples

### Import Required Module

```python
from array import array
```

### Declare Array (Defined Once)

```python
sales = array('i', [25, 30, 28, 35, 40])
print('Sales Array:', sales)
```

## ðŸ”¹ BASIC ARRAY OPERATIONS

### Access Elements

```python
print('First element:', sales[0])
print('Last element:', sales[-1])
```

### Update Element

```python
sales[1] = 32
print('After update:', sales)
```

### Length of Array

```python
print('Length of array:', len(sales))
```

### Append Element

```python
sales.append(45)
print('After append:', sales)
```

### Insert Element

```python
sales.insert(2, 27)
print('After insert:', sales)
```

### Remove Element

```python
sales.remove(27)
print('After remove:', sales)
```

### Pop Element

```python
removed = sales.pop()
print('Popped element:', removed)
print('After pop:', sales)
```

### Index of Element

```python
print('Index of 35:', sales.index(35))
```

### Count Occurrence

```python
print('Count of 30:', sales.count(30))
```

### Reverse Array

```python
sales.reverse()
print('Reversed array:', sales)
```

### Convert Array to List

```python
sales_list = sales.tolist()
print('Converted to list:', sales_list)
```

## ðŸ”¹ ARITHMETIC & LOGICAL PROBLEM STATEMENTS

### Problem 1: Total Sales

```python
print('Total Sales:', sum(sales))
```

### Problem 2: Average Sales

```python
avg = sum(sales) / len(sales)
print('Average Sales:', avg)
```

### Problem 3: Highest and Lowest Sales

```python
print('Highest:', max(sales))
print('Lowest:', min(sales))
```

### Problem 4: Increase Sales by 10%

```python
increased = array('f', [s + s*0.10 for s in sales])
print('After 10% increase:', increased)
```

### Problem 5: Monthly Profit (Expense = 20)

```python
profit = array('i', [s - 20 for s in sales])
print('Monthly Profit:', profit)
```

### Problem 6: Even Sales Values

```python
even = array('i', [s for s in sales if s % 2 == 0])
print('Even Sales:', even)
```

### Problem 7: Square of Sales

```python
square = array('i', [s**2 for s in sales])
print('Square of Sales:', square)
```

### Problem 8: Sales Above Average

```python
above_avg = array('i', [s for s in sales if s > avg])
print('Sales above average:', above_avg)
```

### Problem 9: Difference Between Consecutive Sales

```python
diff = array('i', [sales[i+1] - sales[i] for i in range(len(sales)-1)])
print('Differences:', diff)
```

### Problem 10: Discount Using Floor Division

```python
discount = array('i', [s // 5 for s in sales])
print('Discount:', discount)
```
